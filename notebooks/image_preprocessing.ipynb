# Image Preprocessing

```python
import os
import rasterio
from rasterio.plot import show

import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
tifs = [file for file in os.listdir('data/download/') if file.endswith('.tif')]
tifs
```

```python
img = rasterio.open('data/download/' + tifs[-1], 'r').read(1)
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

```python
plt.hist(img.flatten(), bins=256)
plt.show()
```

```python
def stretch_image(db_img: np.ndarray, min: int, max: int) -> np.ndarray:
    cliped_image = np.clip(db_img, min, max)
    scaled_image = (cliped_image - min) / (max - min)
    # scaled_image = np.nan_to_num(scaled_image)
    # return (scaled_image * 255).astype('uint8')
    return scaled_image
```

```python
sti = stretch_image(img, -30, 0)
```

```python
plt.imshow(sti, cmap='gray')
plt.show()
```

```python
plt.hist(sti.flatten(), bins=256)
plt.show()
```
